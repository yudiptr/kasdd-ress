Berdasarkan artikel berikut [ini](https://www.sciencedaily.com/releases/2015/03/150306102714.htm), dapat diketahui bahwa salah satu faktor yang berpengaruh terhadap precipitation atau hujan adalah kelembapan dari tanah atau soil moisture. Pada dataset harian, tidak terdapat informasi terkait soil moisture, sedangkan pada dataset jam-an ada informasi terkait hal tersebut. Kita akan mencoba untuk menggabungkan kedua dataset tersebut.

```python
# kota unik
cities = list(df_hourly_first.city.unique())

# kolom yang menunjukkan soil moisture
moisture_columns = [col_name
                    for col_name in list(df_hourly_first.columns)
                    if col_name.startswith('soil_moisture')]
print(cities)
print(moisture_columns)
```

```python
def FE_diff_soil(df, shifts):
  # kalau perbedaan soil moisturenya negatif, rescale jadi 0
  def rescale_moisture(x):
    if x < 0 : return 0
    else: return x

  # lakukan dari layer paling bawah, sehingga reverse list
  col = moisture_columns[::-1]

  # weight untuk tiap layer tanah, dalam hal ini berapa jaraknya
  weights = {0: 255-100, 1: 100-28, 2: 28-7, 3: 7-0}

  # dataframe kosong untuk tempat penyimpanan hasil
  df_concat_all = pd.DataFrame(columns=df.columns)

  # lakukan operasi di tiap city
  for city in df.city.unique():
    df_now = df[df['city'] == city]

    for i in range(len(col)):
      acc = 0
      new_col_name = col[i] + '_dif_' + str(shifts)
      if i != 0:
        # akumulasi dari layer yang yang lebih atas
        acc = df_now[col[i-1] + '_dif_' + str(shifts)]

      # mengisi dengan selisih n jam sebelumnya yang ditentukan oleh shift, lalu menerapkan pembobotan
      # ini untuk layer tertentu saja, belum dijumlahkan dengan akumulasi layer-layer di atasnya
      # asumsi mengikuti fenomena semakin lama waktu terakhir hujannya, seharusnya airnya akan semakin
      # meresap ke layer bawah yang dapat meningkatkan moisturenya
      df_now[new_col_name] = (df_now[col[i]] - df_now[col[i]].shift(shifts)).fillna(0) * weights[i]

      # akumulasi layer-layer dan rescale agar tidak minus
      df_now[new_col_name] = (df_now[new_col_name] + acc).apply(rescale_moisture)

    # gabungkan untuk tiap city
    df_concat_all = pd.concat([df_concat_all, df_now], axis=0)

  return df_concat_all
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

from statsmodels.tsa.stattools import adfuller
def evaluate_stationarity(timeseries, t=30):
 #Determing rolling statistics
  aw = timeseries
  rolmean = timeseries
  rolstd = timeseries
  #Plot rolling statistics:
  #orig = plt.plot(timeseries, color='blue',label='Original')
  mean = plt.plot(rolmean, color='red', label='Rolling Mean')
  #std = plt.plot(rolstd, color='black', label = 'Rolling Std')
  plt.legend(loc='best')
  plt.title('Rolling Mean & Standard Deviation')
  plt.show(block=False)
  #Perform Dickey-Fuller test:
  print('Results of Dickey-Fuller Test:')
  dftest = adfuller(timeseries, autolag='AIC')
  dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value','#Lags Used','Number of Observations Used'])
  for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
  print(dfoutput)
```

```python
from pingouin import partial_corr
def partial_correlation(X, y):
  out = pd.Series(index = X.columns, dtype = float)
  for feature_name in X.columns:
    out[feature_name] = partial_corr(
      data = pd.concat([X, y], axis = 1).astype(float),
      x = feature_name,
      y = y.name,
      x_covar = [f for f in X.columns if f != feature_name]
    ).loc['pearson', 'r']
  return out

parshap_test = partial_correlation(shap_test, pd.Series(cbc.predict(feat_test.drop(['rain_sum (mm)']+feat_test.select_dtypes('object').columns.tolist(), axis=1))).rename('rain_sum (mm)'))
parshap_train = partial_correlation(shap_train, feat_train['rain_sum (mm)'])
```
